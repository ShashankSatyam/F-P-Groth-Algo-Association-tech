```python
from mlxtend.frequent_patterns import fpgrowth , association_rules
```

```python
import pandas as pd
df = pd.read_excel('/content/Online Retail.xlsx')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype(str)
df = df[~df['InvoiceNo'].str.contains('C')]
df.head()
```

```python
df['Country'].unique()
```

```python
basket = (df[df['Country'] == 'France'].groupby(['InvoiceNo' , 'Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
basket
```

```python
def encode_unit(x):
  if x<=0:
    return 0
  else:
    return 1
```

```python
basket_set = basket.applymap(encode_unit)
```

```python
basket_set
```

```python
basket_set.drop('POSTAGE' , inplace = True , axis = 1 )
```

```python
freq_itemset = fpgrowth(basket_set , min_support = .1 , use_colnames= True)
```

```python
rules = association_rules(freq_itemset , metric = 'lift')
```

```python
rules
```

```python
# apriori ---------------
```

```python
from mlxtend.frequent_patterns import apriori
freq_itemset = apriori(basket_set , min_support = .1 , use_colnames= True)
```

```python
rules = association_rules(freq_itemset , metric = 'lift')
```
